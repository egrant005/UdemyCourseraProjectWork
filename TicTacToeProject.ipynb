```python
from IPython.display import clear_output
import random
import time
```

```python
def display_board(board,player_marker,player_turn):
    # Display board in current state
    
    marks = [' ',' ',' ',' ',' ',' ',' ',' ',' ']
    nums =  [' ',' ',' ',' ',' ',' ',' ',' ',' ']
    
    for idx in range(0,9):
        if board[idx].isdigit():
            nums[idx] = board[idx]
        else:
            marks[idx] = board[idx]
    
    print(f'Player 1 Marker = {player_marker[0]}')
    print(f'Player 2 Marker = {player_marker[1]}')
    print('')
    print(' {} | {} | {}      {} | {} | {} '.format(marks[6],marks[7],marks[8],nums[6],nums[7],nums[8]))
    print('-----------    -----------')
    print(' {} | {} | {}      {} | {} | {} '.format(marks[3],marks[4],marks[5],nums[3],nums[4],nums[5]))
    print('-----------    -----------')
    print(' {} | {} | {}      {} | {} | {} '.format(marks[0],marks[1],marks[2],nums[0],nums[1],nums[2]))
    print('')
    if player_turn in [1,2]:
        print(f'Player {player_turn}\'s turn')
```

```python
def clear_board():
    # Re-initialize the board
    return ['1','2','3','4','5','6','7','8','9']
```

```python
def mark_board(board,user_selection,user_mark):
    # Mark the board with the user selection
    if user_selection in range(0,9) and board[user_selection] == f'{user_selection+1}':
        board[user_selection] = user_mark
    return board
```

```python
def not_valid_choice(choice,board):
    # Return True is user response is valid - i.e.,
    # choice is a digit between 1 and 9
    # choice is available on the board (i.e., not already taken)
    if  choice.isdigit() and int(choice) in range(1,10) and board[int(choice)-1] == choice:
        return False
    else:
        return True
```

```python
def choice_string(choice,board,player_num):
    # Create a new string to display with input based on what is wrong with initial user input choice
    if  not (choice.isdigit()):
        return f'Player {player_num}'+' - Previous response not a digit. Please enter a numbered location to mark (0 to quit): '
    elif not (int(choice) in range(0,10)):
        return f'Player {player_num}'+' - Previous response outside of range. Please enter a numbered location to mark (0 to quit): '
    elif not (board[int(choice)] == choice):
        return f'Player {player_num}'+' - Previous response not an available location. Please enter a numbered location to mark (0 to quit): '
    else:
        return ''
```

```python
def is_winner(board):
    # Check all possible winning possiblities, return True if a row/col/diag is all X's or O's
    check_idx = [[1,2,3],[4,5,6],[7,8,9],[1,4,7],[2,5,8],[3,6,9],[1,5,9],[7,5,3]]
    for ii in range(0,8):
        sumXO = [0,0]
        for jj in check_idx[ii]:
            if board[jj-1].isdigit():
                break
            elif board[jj-1] == 'X':
                sumXO[0] += 1
            else:
                sumXO[1] += 1
        if 3 in sumXO:
            return True
    return False
```

```python
def computer_player_input(board,player_marker,player_id):
    
    if player_id == 1:
        me = 0
        them = 1
    else:
        me = 1
        them = 0
    
    marks = [' ',' ',' ',' ',' ',' ',' ',' ',' ']
    nums =  [' ',' ',' ',' ',' ',' ',' ',' ',' ']
    
    for idx in range(0,9):
        if board[idx].isdigit():
            nums[idx] = board[idx]
        else:
            marks[idx] = board[idx]    
    
    for idx in range(8,-1,-1):
        if nums[idx] == ' ':
            nums.pop(idx)
    
    
    # Check all possible winning possiblities, return True if a row/col/diag is all X's or O's
    check_idx = [[1,2,3],[4,5,6],[7,8,9],[1,4,7],[2,5,8],[3,6,9],[1,5,9],[7,5,3]]
    
    #TAKE win, if can win
    for ii in range(0,8):
        sumXO = [0,0]
        for jj in check_idx[ii]:
            if board[jj-1].isdigit():
                new_idx = jj
            elif board[jj-1] == player_marker[them]:
                sumXO[them] += 1
            else:
                sumXO[me] += 1
        if sumXO[them] == 0 and sumXO[me] == 2:
            return new_idx
    
    #MAKE sure to block Player 1 from winning
    for ii in range(0,8):
        sumXO = [0,0]
        for jj in check_idx[ii]:
            if board[jj-1].isdigit():
                new_idx = jj
            elif board[jj-1] == player_marker[them]:
                sumXO[them] += 1
            else:
                sumXO[me] += 1
        if sumXO[them] == 2 and sumXO[me] == 0:
            return new_idx
    
    #MAKE sure first mark is either center or corners
    if len(nums) == 9:
        new_idx = random.randint(0,4)
        nums = ['1','3','5','7','9']
        return nums[new_idx]
    
    
    if len(nums) == 8:    
        #IF player 1 goes first and puts mark in center, mark a corner
        if marks[4] == player_marker[them]:
            new_idx = random.randint(0,3)
            nums = ['1','3','7','9']
            return nums[new_idx]
        
        #IF player 1 goes first and puts mark in corner
        for idx in range(0,9):
            if marks[idx] == player_marker[them] and (idx+1) in [1,2,3,4,6,7,8,9]:#[1,3,7,9]:
                return '5'    
   
   
    if len(nums) == 7:
        #IF player 2 goes first (corner) and player 1 marks center, then player 2 marks opposite corner
        if marks[4] == player_marker[them]:
            if marks[0] == player_marker[me]:
                return '9'
            elif marks[2] == player_marker[me]:
                return '7'
            elif marks[6] == player_marker[me]:
                return '3'
            elif marks[8] == player_marker[me]:
                return '1'
            
        #IF player 2 goes first (corner) and player 1 marks a corner, then player 2 marks center
        elif (player_marker[them] in [marks[0],marks[2],marks[6],marks[8]] and
              player_marker[me] in [marks[0],marks[2],marks[6],marks[8]]):
            return '5'
                
    
    if len(nums) == 6:
        #IF player 1 goes first and puts their first 2 marks in opposite corners
        if ((board[0] == player_marker[them] and board[8] == player_marker[them]) or 
           (board[2] == player_marker[them] and board[6] == player_marker[them])):
            new_idx = random.randint(0,3)
            nums = ['2','4','6','8']
            return nums[new_idx]
    
    
        #with open("Output.txt", "w") as text_file:
        #    print(f"Marks: {marks}", file=text_file)
        #    print(f"Marks56: {[marks[5],marks[6]]}", file=text_file)
        #    print(f"Marks56: {[player_marker[0]]*2}", file=text_file)
        #    print(f"Marks56: {[marks[5],marks[6]] == [player_marker[0]]*2}", file=text_file)
    
        #IF player 1 goes first and puts their first mark in a corner and second in an edge
        if ([marks[0],marks[7]] == [player_marker[them]]*2 or [marks[3],marks[8]] == [player_marker[them]]*2):
            return '7'
        elif ([marks[0],marks[5]] == [player_marker[them]]*2 or [marks[1],marks[8]] == [player_marker[them]]*2):
            return '3'
        elif ([marks[2],marks[7]] == [player_marker[them]]*2 or [marks[5],marks[6]] == [player_marker[them]]*2):
            return '9'
        elif ([marks[2],marks[3]] == [player_marker[them]]*2 or [marks[1],marks[6]] == [player_marker[them]]*2):
            return '1'
    
    if len(nums) == 5:
        #If player 2 goes first marking opposite corners and player 1 marks center and a corner
        if [marks[0],marks[8]] == [player_marker[me]]*2:
            if [marks[5],marks[2]] == [player_marker[them]]*2:
                return '7'
            elif [marks[5],marks[6]] == [player_marker[them]]*2:
                return '3'
        elif [marks[2],marks[6]] == [player_marker[me]]*2:
            if [marks[5],marks[0]] == [player_marker[them]]*2:
                return '9'
            elif [marks[5],marks[8]] == [player_marker[them]]*2:
                return '1'
        
        
        #If player 2 goes first marking a corner and center, and player 1 goes opposite corner and edge
        if (([marks[0],marks[4]] == [player_marker[me]]*2 and marks[8] == player_marker[them]) or
            ([marks[4],marks[8]] == [player_marker[me]]*2 and marks[0] == player_marker[them])):
            if player_marker[them] in [marks[1],marks[5]]:
                return '6'
            elif player_marker[them] in [marks[3],marks[7]]:
                return '2'
        elif (([marks[2],marks[4]] == [player_marker[me]]*2 and marks[6] == player_marker[them]) or
             ([marks[4],marks[6]] == [player_marker[me]]*2 and marks[2] == player_marker[them])):
            if player_marker[them] in [marks[1],marks[3]]:
                return '8'
            elif player_marker[them] in [marks[5],marks[7]]:
                return '1'
    
    #ELSE, just add mark to win next turn
    for ii in range(0,8):
        sumXO = [0,0]
        for jj in check_idx[ii]:
            if board[jj-1].isdigit():
                new_idx = jj
            elif board[jj-1] == player_marker[them]:
                sumXO[them] += 1
            else:
                sumXO[me] += 1
        if sumXO[them] == 0 and sumXO[me] == 1:
            return new_idx
    
    new_idx = random.randint(0,(len(nums)-1))
    
    return nums[new_idx]
```

```python
def main_game():
    
    clear_output()
    board = clear_board()
    player_wins = [0,0,0]

    choice = input('Welcome to Tic-Tac-Toe\nWould you like to play a game (Y/N)? ')
    if 'Y' in choice.upper():        

        # Computer or Human Player 2
        clear_output()
        choice = input('1 Player or 2 Player game (1/2)? ')
        if '1' in choice:
            num_players = 1
        elif '2' in choice:
            num_players = 2
        else:
            num_players = 0
        
        # Choose markers for Player 1 and 2
        clear_output()
        choice = input('Choose a marker for Player 1 (X/O): ')
        if 'X' in choice.upper():
            player_marker = ['X','O']
        else:
            player_marker = ['O','X']

        clear_output()
        
        #Initialize some variables
        play_game = True   # Bool to continue gameplay             
        player_first = random.randint(1,2) # Keeps track of who goes first for each game (alternates)
        player_turn = player_first # Keeps track of whos turn it is
        turn_count = 0
        
        while play_game:

            # Player Turns
            #################        
            clear_output()
            display_board(board,player_marker,player_turn)

            if num_players == 2 or (num_players == 1 and player_turn == 1):
                time.sleep(0.5)
                choice = input(f'Player {player_turn}  - Please enter a numbered location to mark (0 to quit): ')
                if choice.isdigit() and int(choice) == 0:
                        play_game = False
                        break
                else:
                    while not_valid_choice(choice,board):
                        choice = input(choice_string(choice,board,player_turn))
                        if choice.isdigit() and int(choice) == 0:
                            play_game = False
                            break
            else:
                choice = computer_player_input(board,player_marker,player_turn)
           
            # Update board with player selection
            # Check to see if a winner/cat's game
            # Update player wins OR draws
            # Update first player if new game wanted
            #####################################
            if play_game: 
                turn_count += 1
                board = mark_board(board,int(choice)-1,player_marker[player_turn-1])        
                if is_winner(board) or turn_count == 8:
                    
                    if turn_count == 8 and not(is_winner(board)):
                        player_turn = player_turn%2 + 1
                        for idx in range(0,9):
                            if board[idx].isdigit():
                                board[idx] = player_marker[player_turn-1]
                                
                    clear_output()
                    display_board(board,player_marker,0)
                    
                    if is_winner(board):
                        print(f'Player {player_turn} wins!')
                        print(f'Player {player_turn} wins!')
                        print(f'Player {player_turn} wins!')
                        player_wins[player_turn-1] += 1
                    else:
                        print('Cat\'s Game!')
                        print('Cat\'s Game!')
                        print('Cat\'s Game!')
                        player_wins[2] += 1
                        
                    player_first = player_first%2 + 1
                    
                    time.sleep(0.5)
                    choice = input('Would you like to play another game (Y/N)? ')
                    if 'Y' in choice.upper(): 
                        clear_output()
                        board = clear_board()
                        turn_count = 0
                    else:
                        play_game = False
                
                # Change player turn based on if new game or continuation of current game
                if turn_count == 0:
                    player_turn = player_first
                else:
                    player_turn = player_turn%2 + 1
        
    clear_output()
    print('')
    print('Game over')
    print('')
    print(f'Player 1 wins = {player_wins[0]}')
    print(f'Player 2 wins = {player_wins[1]}')
    print(f'Draws = {player_wins[2]}')
    print('')
    print('Goodbye!')
    print('')
```
