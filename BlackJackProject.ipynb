```python
# Create N shuffled deck(s) 
def Deck(num_decks=1, rand_seed=''):    
        
    card_deck = ['']*52*num_decks

    card_counter = -1
    for deck in range(0,num_decks):
        for suit in suits:
            for rank in ranks:
                card_counter += 1
                card_deck[card_counter] = f'{rank}{suit}'
    
    if rand_seed != '':
        random.seed(rand_seed + pass_num*10010)
    random.shuffle(card_deck)    
    
    return(card_deck)
        
```

```python
# Define the BlackJack Card Game
class BlackJack():
    
    def __init__(self, num_decks=6, depth_before_shuffle=0.66, dealer_hits_soft_17=False, split_after_split = True, double_down=True,
                  double_down_after_split=True, early_surrender=False, late_surrender=True, seven_card_charlie=True, min_bet = 50, max_bet=500, payoff_ratio=3/2):
        
        self.num_decks = num_decks
        self.depth_before_shuffle = depth_before_shuffle
        self.dealer_hits_soft_17  = dealer_hits_soft_17
        self.split_after_split = split_after_split
        self.double_down  = double_down
        self.double_down_after_split  = double_down_after_split
        self.early_surrender  = early_surrender
        self.late_surrender  = late_surrender
        self.seven_card_charlie = seven_card_charlie
        self.min_bet = min_bet
        self.max_bet = max_bet
        
        # Define pay-off ratio depending on deck count used in game. Lower the deck count, lower the pay-out
        self.payoff_ratio = payoff_ratio
        if ((payoff_ratio-3/2)**2)**0.5 < 0.01:
            self.por_str = '3/2'
        elif ((payoff_ratio-5/4)**2)**0.5 < 0.01:
            self.por_str = '5/4'
        elif ((payoff_ratio-6/5)**2)**0.5 < 0.01:
            self.por_str = '6/5'
        elif ((payoff_ratio-1)**2)**0.5 < 0.01:
            self.por_str = '1/1'
    
    # Generate new game with randomly determined game properties
    def new_random_game(self):
        
        ## Randomly select number of decks
        self.num_decks = random.sample(range(1, 9), 1)[0]

        ## Randomly determine depth of total deck(s) before shuffling
        self.depth_before_shuffle = 0.75 - random.sample(range(0,100),1)[0]/100*0.74

        ## Randomly determine if dealer hits or stays on soft 17
        if random.sample(range(0,2),1)[0] == 1:
            self.dealer_hits_soft_17 = True
        else:
            self.dealer_hits_soft_17 = False

        ## Randomly determine if player can split after a split
        if random.sample(range(0,2),1)[0] == 1:
            self.split_after_split = True
        else:
            self.split_after_split = False
            
        ## Randomly allow double downs and double down after split
        if random.sample(range(0,2),1)[0] == 1:
            self.double_down = True
            if random.sample(range(0,2),1)[0] == 1:
                self.double_down_after_split = True
            else:
                self.double_down_after_split = False
        else: 
            self.double_down = False
            self.double_down_after_split = False

        ## Randomly allow early/late surrender
        if random.sample(range(0,2),1)[0] == 1:
            self.early_surrender = False
            if random.sample(range(0,2),1)[0] == 1:
                self.late_surrender = True
            else:
                self.late_surrender = False
        else: 
            self.early_surrender = True
            self.late_surrender = True

        ## Randomly allow Seven Card Charlie
        if random.sample(range(0,2),1)[0] == 1:
            self.seven_card_charlie = True
        else:
            self.seven_card_charlie = False

        # Randomly determine max bet     
        self.min_bet = 5 * 10**random.sample(range(0,3),1)[0]
        self.max_bet = self.min_bet * 10

        # payoff based on game setup
        # Currently just based on number of decks
        # Need to adjust based on favorability to Player/House
        if self.num_decks == 1:
            self.payoff_ratio = 6/5
            self.por_str = '6/5'
        elif self.num_decks == 2:
            self.payoff_ratio = 5/4
            self.por_str = '5/4'
        else:
            self.payoff_ratio = 3/2 
            self.por_str = '3/2'
    
    # Print the rules of the game
    def print_rules(self):
        print('')
        print('Welcome to BlackJack - Please read the house rules below:')
        print(f'\t- Dealer will deal from {self.num_decks} deck(s)')
        print(f'\t- Dealer will reshuffle after {int(self.depth_before_shuffle*100)}% cards dealt from the {self.num_decks*52} card deck')

        temp_str='NOT '
        if self.dealer_hits_soft_17:
            temp_str = ''
        print(f'\t- Dealer will {temp_str}hit on a "soft 17"')

        temp_str='NOT '
        if self.split_after_split:
            temp_str = ''
        print(f'\t- Splitting after a split is {temp_str}allowed')
        
        temp_str='NOT '
        if self.double_down:
            temp_str = ''
        print(f'\t- Double downs are {temp_str}allowed')

        if self.double_down: 
            temp_str1='But'
            temp_str2='NOT '
            if self.double_down_after_split:
                temp_str1='Also'
                temp_str2=''
            print(f'\t- {temp_str1} double downs after splitting are {temp_str2}allowed')

        if self.early_surrender:
            print('\t- Early AND Late surrenders are allowed')
        elif self.late_surrender: 
            print('\t- Early surrenders are NOT allowed, but Late surrenders are allowed.')
        else:
            print('\t- Early AND Late surrenders are NOT allowed')

        temp_str='NOT '
        if self.seven_card_charlie:
            temp_str = ''
        print(f'\t- Seven card Charlie is {temp_str}allowed')

        print(f'\t- The minimum player bet is {self.min_bet}')
        print(f'\t- The maximum player bet is {self.max_bet}')
        print(f'\t- The pay-out ratio is {self.por_str}')
        print('')
        time.sleep(0.5)
```

```python
# Create Player profile and actions

class Player():
    
    # If human player = False (i.e., a Computer player)
    # can set optional "nature" parameter to indicate aggressiveness
    # - Conservative
    # - Dealer
    # - Aggressive
    
    def __init__(self, name, chip_count=0, human=True, blackjack_setup=BlackJack()):        
        
        self.name = name
        self.chip_count = chip_count
        self.human = human
        self.blackjack_setup = blackjack_setup
        
        self.wins_in_a_row = 0
        self.record_for_last10 = []
        self.record_for_last10_sum = 0
        
        self.computer_runs = 0
        self.computer_run_limit = 10000
        self.starting_chip_count = 0
        
        self.clear_count()        
        self.reset()
        
    # Take card and append to hand index indicated by "hand_idx"
    # Calculate possible values for each hand index
    # Determine if hand is "splittable"
    def take_card(self,new_card,hand_idx=0):
        
        # Append a card to Hand "hand_idx" and initialize hand property variables  
        self.hand[hand_idx].append(new_card)
        self.hand_count = len(self.hand)

        self.hand_value = [[0,0]]
        for idx in range(1,self.hand_count):                
            self.hand_value.append([0,0])
        self.hand_splittable = [False]*self.hand_count
            
            
        # Determine hand values for each hand
        for hand_idx in range(0,self.hand_count):
            num_aces = 0
            for card in self.hand[hand_idx]:
                self.hand_value[hand_idx][0] += values[card[0:2]][0]
                self.hand_value[hand_idx][1] += values[card[0:2]][1]
                if card[0:2] == 'Ac':
                    num_aces += 1

            # Reduce largest hand value amount if hand includes more than 1 ace
            if num_aces > 1.1:
                self.hand_value[hand_idx][1] = self.hand_value[hand_idx][1] - 10*(num_aces-1)  

            # Determine if hand is splittable
            if len(self.hand[hand_idx]) == 2:
                if self.hand[hand_idx][0][0:2] == self.hand[hand_idx][1][0:2]:
                    self.hand_splittable[hand_idx] = True
    
    # Reset hand and betting information for each new deal
    def reset(self):
        self.hand = [[]]
        self.hand_count = 1
        self.hand_value = [[0,0]]
        self.hand_splittable = [False]        
        self.bet_amount = [0]
    
    # Split current hand and place equal bet on newly split off hand
    def split_hand(self,hand_idx=0):
        
        #Append new hand to player, assign to second card from hand_idx, and
        # then pop off the second card from hand_idx
        self.hand.append([self.hand[hand_idx][1]])
        self.hand[hand_idx].pop()

        # Copy original bet amount into split off hand
        self.bet_amount.append(0)
        self.bet(self.bet_amount[hand_idx],hand_idx+1)

        # Initialize newly split hand as unspittable
        self.hand_splittable.append(False)
        self.hand_splittable[hand_idx] = False

        # Update hand count
        self.hand_count = len(self.hand)
        
        # Initialize hand values
        self.hand_value = [[0,0]]
        for idx in range(1,self.hand_count):                
            self.hand_value.append([0,0])
            
        # Populate new hand values
        for hand_idx in range(0,self.hand_count):
            for card in self.hand[hand_idx]:
                self.hand_value[hand_idx][0] += values[card[0:2]][0]
                self.hand_value[hand_idx][1] += values[card[0:2]][1]    
    
    # Double bet on current hand and take only 1 more card
    def double_down(self,new_card,hand_idx=0):
        self.bet(self.bet_amount[hand_idx],hand_idx)        
        self.take_card(new_card,hand_idx)
    
    # Set bet amount & subtracted from chip count
    def bet(self, amount, hand_idx=0):
        self.bet_amount[hand_idx] = self.bet_amount[hand_idx]+amount        
        self.chip_count -= amount     
    
    def computer_bet(self, previous_bet=0):
        
        '''
        # conservative betting when card_count = 1 - 8
        # neutral betting when card_count = 9 - 13
        # aggressive bettern when card_count = 14+
        
        
        if self.card_count < 2:
            bet_temp1 = self.blackjack_setup.min_bet
        elif self.card_count < 9:
            bet_temp1 = 2*self.blackjack_setup.min_bet
        elif self.card_count < 13:
            bet_temp1 = int( (self.blackjack_setup.max_bet-self.blackjack_setup.min_bet)/2 )
        else:
            bet_temp1 = self.blackjack_setup.max_bet   
        '''
        
        # Define various betting strategies to use for testing
        temp_max = min([self.blackjack_setup.max_bet,self.chip_count])
        
        if strategy_num == 1: #Progressive
            return min([self.blackjack_setup.min_bet * (self.wins_in_a_row+1), temp_max])
        elif strategy_num == 2: # Hybrid            
            
            if self.record_for_last10_sum > 5:
                if self.wins_in_a_row == 0:
                    return self.blackjack_setup.min_bet*2
                elif self.wins_in_a_row < 2:
                    return self.blackjack_setup.min_bet*2
                elif self.wins_in_a_row < 4:
                    return self.blackjack_setup.min_bet*4
                elif self.wins_in_a_row < 6:
                    return self.blackjack_setup.min_bet*4
                elif self.wins_in_a_row < 8:
                    return self.blackjack_setup.min_bet*8
                else:
                    return self.blackjack_setup.min_bet*8
            else:
                return self.blackjack_setup.min_bet            
                
        else: # Min only
            return self.blackjack_setup.min_bet
    
    def computer_play(self, dealer_card, hand_idx=0, play_str='H/S', play_like_dealer=False):        
        #based on house rules
        #based on programed strategy 
        #based on card_count            
        
        # H = (H)it
        # S = (S)tand
        # P = s(P)lit
        # D = (D)ouble down
        # R = su(R)render
        # A = pl(A)y
        
        if play_like_dealer:
            if self.hand_value[hand_idx][0] < 17 and not(self.hand_value[hand_idx][1] >= 17 and self.hand_value[hand_idx][1] <= 21):
                return 'H'
            else:
                return 'S'
        else:
            if len(self.hand[hand_idx]) == 1:
                return 'H'
                
            num_aces = 0
            for card in self.hand[hand_idx]:
                if card[0:2] == 'Ac':
                    num_aces += 1
                    
            if self.hand_value[hand_idx][0] == 21 or self.hand_value[hand_idx][1] == 21:
                return 'S'
            elif self.hand_splittable[hand_idx]:
                #Use SPLIT Strategy
                temp_str = strategy_split[(self.hand_value[hand_idx][0],values[dealer_card[0:2]][1])]
            elif ((self.hand_value[hand_idx][0] <= 21 and self.hand_value[hand_idx][1] > 21) or
                   (self.hand_value[hand_idx][0] <= 21 and self.hand_value[hand_idx][0] == self.hand_value[hand_idx][1])):
                #Use HARD Strategy
                if self.hand_value[hand_idx][0] < 8:
                    return 'H'
                elif 8 <= self.hand_value[hand_idx][0] <= 17:
                    temp_str = strategy_hard[(self.hand_value[hand_idx][0],values[dealer_card[0:2]][1])]
                else:
                    return 'S'
            else:
                #Use SOFT strategy
                temp_str = strategy_soft[(self.hand_value[hand_idx][0],values[dealer_card[0:2]][1])]
            
            if len(temp_str) == 1:            
                return temp_str
            else:
                for idx in range(0,2):
                    if temp_str[idx] in play_str:
                        return temp_str[idx]
                return 'A'
    
    # Card counting - adjust count according to card values
    def adjust_count(self, card_value, num_decks):
        # conservative betting when card_count = 1 - 8
        # neutral betting when card_count = 9 - 13
        # aggressive bettern when card_count = 14+        
        if card_value < 7:
            self.card_count += 1
        elif card_value > 9:
            self.card_count -= 1
        
        self.card_count_per_deck = int(self.card_count/num_decks)
        
    # Clear card count after each shuffle
    def clear_count(self):
        self.card_count = 0
        self.card_count_per_deck = 0         
    
    # Function for adjusting a list storing a running counter for wins/losses
    def adjust_record_and_sum(self, num_temp):
        self.record_for_last10.append(num_temp)
        if len(self.record_for_last10) > 10:
            self.record_for_last10.reverse()
            self.record_for_last10.pop()
            self.record_for_last10.reverse()
        
        self.record_for_last10_sum = 0        
        for num in self.record_for_last10:
            self.record_for_last10_sum += num
            
        if num_temp > 0.1:
            if self.wins_in_a_row >= 0:
                self.wins_in_a_row += 1
            else:
                self.wins_in_a_row = 1
        elif num_temp < -0.1:
            if self.wins_in_a_row <= 0:
                self.wins_in_a_row -= 1
            else:
                self.wins_in_a_row = -1
        else:
            self.wins_in_a_row = 0
    
    # Determine if all hands surrendered
    def all_surrendered(self):
        for hand_idx in range(0,self.hand_count):
            if self.hand_value[hand_idx][0] > 0:
                return False
        return True
    
    # Determine if all hands busted
    def all_busted(self):
        for hand_idx in range(0,self.hand_count):
            if 0 < self.hand_value[hand_idx][0] <= 21:
                return False
        return True
     
    # Temporary output string for Player class (for debugging)
    def __str__(self):
        temp_str = (f"\nPlayer Name =        {self.name}\n" +
                    f"Player Chip Count =  {self.chip_count}\n" +
                    f"Player is Human =    {self.human}\n")
        
        if len(self.hand) < 1:
            temp_str = temp_str + "Player Holding No Cards!\n"
        elif type(self.hand[0]) == str:
            temp_str = temp_str + "Player Holding Cards:\n"
            for card in self.hand:
                temp_str = temp_str + "\t" + card + "\n"
            temp_str = temp_str + f"Total value of Hand: {self.hand_value[0]}\n"    
        else:
            temp_str = temp_str + "Player Holding Cards:\n"
            for hand_idx in range(0,len(self.hand)):
                temp_str = temp_str + f"\tHand {hand_idx+1}):\n"
                for card in self.hand[hand_idx]:
                    temp_str = temp_str + "\t\t" + card + "\n"
                temp_str = temp_str + f"\tTotal value of Hand {hand_idx+1}: {self.hand_value[hand_idx]}\n"            

        return temp_str
```

```python
# Create Dealer actions

class Dealer():
    
    def __init__(self, blackjack_setup):
        self.blackjack_setup = blackjack_setup
        
        self.hand = []
        self.hand_value = [[0,0]]
        self.show_card = ''
        
        self.new_deck = Deck()
    
    # deal a card from deck to player (Assume Last card = top of deck)
    def deal_card(self):        
        return self.new_deck.pop()
    
    # Append new card to dealers hand
    def take_card(self):
        # dealer hand
        self.hand.append(self.new_deck.pop())
        self.hand_value = [[0,0]]
        num_aces = 0
        for card in self.hand:
            self.hand_value[0][0] += values[card[0:2]][0]
            self.hand_value[0][1] += values[card[0:2]][1]
            if card[0:2] == 'Ac':
                num_aces += 1

        # Reduce largest hand value amount if hand includes more than 1 ace
        if num_aces > 1.1:
            self.hand_value[0][1] = self.hand_value[0][1] - 10*(num_aces-1) 
    
        self.show_card = self.hand[0]
    
    # remove cards from table
    def reset(self):        
        self.hand = []
        self.hand_value = [[0,0]]
        self.show_card = ''
    
    # pay bets to player (if applicable)
    def payoff_bets(self, initial_bet, blackjack=False, surrender=False):        
        if blackjack:
            return int(initial_bet*(1+self.blackjack_setup.payoff_ratio)*100)/100
        elif surrender:
            return int(initial_bet*0.5*100)/100
        else:
            return int(initial_bet*2*100)/100
    
    # generate a new shuffled deck when penetration depth gets to certain point
    def reset_deck(self, rand_seed=''):        
        self.new_deck = Deck(self.blackjack_setup.num_decks, rand_seed)    
    
    # Temporary output string for Dealer class (for debugging)
    def __str__(self):

        if len(self.hand) > 0:
            temp_str = "\nDealer Holding Cards:\n"
            for card in self.hand:
                temp_str = temp_str + "\t" + card + "\n"
            temp_str = temp_str + f"Total value of Hand: {self.hand_value[0]}\n"
        else:
            temp_str = temp_str + "Dealer Holding No Cards!\n"
            
        return temp_str
        
    
```
